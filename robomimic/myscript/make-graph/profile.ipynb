```python
import pandas as pd

# 各タスクのprofile.csvを読み込む
profile_paths = {
    "can": "/work/robomimic/csv/eval/can/profile.csv",
    "lift": "/work/robomimic/csv/eval/lift/profile.csv",
    "square": "/work/robomimic/csv/eval/square/profile.csv",
}

profiles = {}
for task, path in profile_paths.items():
    try:
        profiles[task] = pd.read_csv(path)
        print(f"Loaded: {task} ({path}) shape={profiles[task].shape}")
    except Exception as e:
        print(f"Failed to load {task} ({path}): {e}")

# profiles["can"], profiles["lift"], profiles["square"] で各DataFrameにアクセスできます
```

```python
# 各タスクのprofile.csvをモデルごと（seed1~3は平均化）で表示
import re
def normalize_model_name(name):
    # seed番号を除去（例: ncp_u128_best_seed1 → ncp_u128_best）
    return re.sub(r'_seed[0-9]+$', '', name)

for task, df in profiles.items():
    print(f'--- {task} ---')
    df = df.copy()
    df['model_group'] = df['name'].apply(normalize_model_name)
    means_by_model = df.groupby('model_group').mean(numeric_only=True)
    display(means_by_model)
```
